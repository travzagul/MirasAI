```python
import pandas as pd

# 1. Verilerin Yüklenmesi
# Farklı kaynaklardan (Açık veri portalları ve Kaggle) elde edilen veri setleri okutuluyor.
df1 = pd.read_csv('istanbul_turistik_yerler.csv')
df2 = pd.read_csv('ibb-sosyal-tesis-konumlar.csv', sep=';')
df3 = pd.read_csv('sightseeing_places_ISTANBUL_cleaned.csv')

# 2. Şema Standardizasyonu ve Etiketleme
# Farklı formatlardaki veri setlerinin sistem entegrasyonu için ortak bir şemaya ('Mekan_Adi', 'Enlem', 'Boylam', 'Kategori') dönüştürülmesi işlemi.
df1 = df1[['name', 'latitude', 'longitude']].rename(columns={'name': 'Mekan_Adi', 'latitude': 'Enlem', 'longitude': 'Boylam'})
df1['Kategori'] = 'Turistik Yer'

df2 = df2[['TESİS ADI', 'Latitude', 'Latitude.1']].rename(columns={'TESİS ADI': 'Mekan_Adi', 'Latitude': 'Enlem', 'Latitude.1': 'Boylam'})
df2['Kategori'] = 'İBB Sosyal Tesis'

# İBB koordinat verilerindeki tip uyuşmazlığının (eksik ondalık ayırıcı) string manipülasyonu ile giderilip tekrar float tipine dönüştürülmesi.
df2['Enlem'] = df2['Enlem'].astype(str).apply(lambda x: float(x[:2] + '.' + x[2:]))
df2['Boylam'] = df2['Boylam'].astype(str).apply(lambda x: float(x[:2] + '.' + x[2:]))

df3 = df3[['NAME', 'LATITUDE', 'LONGITUDE']].rename(columns={'NAME': 'Mekan_Adi', 'LATITUDE': 'Enlem', 'LONGITUDE': 'Boylam'})
df3['Kategori'] = 'Turistik Yer'

# 3. Verilerin Birleştirilmesi (Data Merging) ve Kopya Temizliği
# Tekilleştirme işlemi (Drop Duplicates) yapılarak modelin aynı mekanı iki kez önermesinin önüne geçilmiştir.
master_df = pd.concat([df1, df2, df3], ignore_index=True)
master_df = master_df.drop_duplicates(subset=['Mekan_Adi'], keep='first')

# 4. Üretime Hazır (Production-Ready) Verinin Kaydedilmesi
master_df.to_csv('master_lokasyonlar.csv', index=False)

print(f"Veri ön işleme tamamlandı. Toplam {len(master_df)} benzersiz mekan sisteme tanıtıldı.")
```

```python
import numpy as np

def haversine_mesafe(enlem1, boylam1, enlem2, boylam2):
    """
    Dünya düz bir zemin olmadığı için standart Öklid uzaklığı (düz çizgi) formülü İstanbul haritasında hata payı yaratır.
    Bu nedenle, küresel trigonometri prensiplerine dayanan Haversine formülü kullanılarak iki GPS noktası
    arasındaki gerçek "kuş uçuşu" mesafe (km cinsinden) hesaplanmıştır.
    """
    R = 6371.0 # Dünyanın ortalama yarıçapı
    enlem1, boylam1, enlem2, boylam2 = map(np.radians, [enlem1, boylam1, enlem2, boylam2])
    d_enlem = enlem2 - enlem1
    d_boylam = boylam2 - boylam1
    a = np.sin(d_enlem / 2)**2 + np.cos(enlem1) * np.cos(enlem2) * np.sin(d_boylam / 2)**2
    return R * (2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a)))

def akilli_yakin_yerleri_oner(secilen_mekan_adi, df):
    """
    Kullanıcı deneyimini (UX) optimize etmek için tasarlanmış karar destek fonksiyonudur.
    Kullanıcıya mantıksız (çok uzak) öneriler sunmamak adına 3 km'lik bir erişilebilirlik sınırı
    (radius) belirlenmiştir. Eğer bu çapta mekan yoksa, sistem fallback (geri çekilme) senaryosunu
    devreye sokarak en yakın 3 lokasyonu getirir.
    """
    secilen_veri = df[df['Mekan_Adi'] == secilen_mekan_adi]

    if secilen_veri.empty:
        return "Hata: Sistemde bulunmayan bir mekan sorgulandı."

    secilen_enlem = secilen_veri.iloc[0]['Enlem']
    secilen_boylam = secilen_veri.iloc[0]['Boylam']

    # Tüm mekanlara olan mesafelerin dinamik hesaplanması
    df_kopya = df.copy()
    df_kopya['Mesafe_km'] = df_kopya.apply(lambda row: haversine_mesafe(secilen_enlem, secilen_boylam, row['Enlem'], row['Boylam']), axis=1)
    df_kopya = df_kopya[df_kopya['Mekan_Adi'] != secilen_mekan_adi]

    # 3 km kısıtı ve sıralama işlemi
    yakin_yerler = df_kopya[df_kopya['Mesafe_km'] <= 3.0].sort_values(by='Mesafe_km')

    # Karar mekanizması
    if not yakin_yerler.empty:
        return yakin_yerler[['Mekan_Adi', 'Kategori', 'Mesafe_km']]
    else:
        return df_kopya.sort_values(by='Mesafe_km').head(3)[['Mekan_Adi', 'Kategori', 'Mesafe_km']]
```

# BÖLÜM 2: BİLGİSAYARLI GÖRÜ (COMPUTER VISION) VE VERİ HAZIRLIĞI

Sistemin Arka Plan (Backend) ve Tavsiye Motoru geliştirme süreçleri başarıyla tamamlanmıştır. Bu bölümde, projenin temel yapay zeka modülü olan **Görüntü Sınıflandırma** modelinin eğitimi için veri hazırlık (Data Preprocessing) adımlarına geçilmektedir.

Modelin İstanbul'un ikonik yapılarını tanıyabilmesi için takım çalışmasıyla toplanan ham fotoğraflar, doğrudan bu aşamada PyTorch kütüphanesi kullanılarak işlenecektir. Veri setinin kısıtlı kalmaması ve modelin genelleme yeteneğinin (accuracy) artırılması amacıyla, resimler üzerinde dinamik Veri Artırımı (Data Augmentation) teknikleri uygulanacaktır.
---


```python
from google.colab import drive
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import os

# 1. Veri Kaynağına  Bağlantı Kurulması
veri_yolu = '/content/MirasAI_Dataset' # Kendi klasör yoluna göre güncelleyebilirsin

# 2. Veri Artırımı (Data Augmentation) Stratejisi
# Modelin farklı senaryolarda (farklı ışık, amatör çekim açısı vs.) sağlam (robust) çalışabilmesi için
# fotoğraflara rastgele döndürme, yatay çevirme ve parlaklık değişimleri uygulanır.
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# 3. Veri Yükleyici (DataLoader) Konfigürasyonu
dataset = datasets.ImageFolder(root=veri_yolu, transform=train_transforms)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

print("-" * 50)
print(f"Toplam Sınıf (Mekan) Sayısı: {len(dataset.classes)}")
print(f"Eğitime Hazır Toplam Fotoğraf Sayısı: {len(dataset)}")
print("-" * 50)
```

# 🧠 BÖLÜM 3: DERİN ÖĞRENME VE MODEL EĞİTİMİ (TRANSFER LEARNING)

Veri seti hazırlık sürecinin ardından, projenin temel yapay zeka çekirdeği olan **Görüntü Sınıflandırma** aşamasına geçilmiştir. Sistem optimizasyonu ve yüksek doğruluk (accuracy) hedefleri doğrultusunda, modelin sıfırdan eğitilmesi yerine literatürde kendini kanıtlamış **ResNet-18** mimarisi tercih edilmiştir.

Bu bölümde uygulanan **Transfer Learning (Öğrenme Transferi)** ve **Fine-Tuning (İnce Ayar)** adımlarında mühendislik yaklaşımı olarak şu stratejiler izlenmiştir:
1. Modelin milyonlarca nesne üzerinde kazandığı engin görsel hafıza projeye entegre edilmiştir.
2. Sadece son karar katmanı (fully connected layer) kesilmiş ve MirasAI projesinin 12 ikonik mekanını tanıyacak şekilde yeniden yapılandırılmıştır.
3. Donanım kaynaklarının verimli kullanımı için GPU optimizasyonu sağlanmıştır.
---


```python
# ---------------------------------------------------------
# ADIM 1: Kütüphanelerin Yüklenmesi
# ---------------------------------------------------------
from google.colab import drive
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from sklearn.metrics import f1_score, mean_squared_error
import numpy as np
import os

# ---------------------------------------------------------
# --- ADIM 2: Veri Yükleme ve Train/Test Ayrımı ---
full_dataset = datasets.ImageFolder(root='/content/MirasAI_Dataset', transform=train_transforms)

train_size = int(0.8 * len(full_dataset))
test_size = len(full_dataset) - train_size
train_dataset, test_dataset = torch.utils.data.random_split(full_dataset, [train_size, test_size])

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"✅ Sistem Hazır: {len(train_dataset)} eğitim, {len(test_dataset)} test fotoğrafı ayrıldı.")

# ---------------------------------------------------------
# ADIM 3: İLERİ SEVİYE FINE-TUNING (KISMİ KİLİT AÇMA)
# ---------------------------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🚀 Gelişmiş Eğitim şu cihazda başlıyor: {device.type.upper()}")

model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)

# Önce tüm ağın kilidini kapatıyoruz (Transfer Learning)
for param in model.parameters():
    param.requires_grad = False

# STRATEJİ DEĞİŞİKLİĞİ: Sadece son görme bloğunun (layer4) kilidini açıyoruz!
# Bu sayede İstanbul mimarisine has kıvrımları öğrenmeye başlayacak.
for param in model.layer4.parameters():
    param.requires_grad = True

# Çıkış katmanını 12 sınıfımıza göre ayarlıyoruz
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(full_dataset.classes))
model = model.to(device)

criterion = nn.CrossEntropyLoss()

# Öğrenme Oranları (Learning Rate):
# Önceden eğitilmiş layer4'ü çok bozmamak için düşük (1e-4),
# sıfırdan kurduğumuz son katman (fc) için normal (1e-3) hızda eğitiyoruz.
optimizer = optim.Adam([
    {'params': model.layer4.parameters(), 'lr': 1e-4},
    {'params': model.fc.parameters(), 'lr': 1e-3}
], weight_decay=1e-4)
# ---------------------------------------------------------
# ADIM 4: Derin Öğrenme Eğitim Döngüsü
# ---------------------------------------------------------
num_epochs =10 # 5'ten 10'e çıkardık, model daha fazla pratik yapacak
print("-" * 50)
print(f"🧠 İleri Seviye Model Eğitimi Başladı ({num_epochs} Epoch)...")

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    # Test verisiyle doğrulama
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = 100 * correct / total
    print(f"Epoch {epoch+1}/{num_epochs} | Eğitim Hatası: {epoch_loss:.4f} | Test Doğruluğu: %{epoch_acc:.2f}")

# ---------------------------------------------------------
# EK ADIM: Performans Metrikleri
# ---------------------------------------------------------
all_preds = []
all_labels = []

model.eval()
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())

f1 = f1_score(all_labels, all_preds, average='weighted')
mse = mean_squared_error(all_labels, all_preds)

print("-" * 50)
print(f"📊 Final Performans Metrikleri:")
print(f"✅ F1-Score: {f1:.4f}")
print(f"📉 MSE: {mse:.4f}")
print("-" * 50)

# ---------------------------------------------------------
# ADIM 5: Model Ağırlıklarının Kaydedilmesi
# ---------------------------------------------------------
torch.save(model.state_dict(), 'miras_ai_resnet18_weights.pth')
print("🎯 İleri seviye eğitim tamamlandı! Yeni ağırlıklar kaydedildi.")
```
